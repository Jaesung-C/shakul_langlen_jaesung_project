```python
import numpy as np
import matplotlib.pyplot as plt

system_type = 'SSS' 
V = 10**4
parm_idx = 2 
attribute = 'x'
sample_idx = 2 

base_dir = f'./data_set/{system_type}/euler/V{V}/parm{parm_idx}/'

sample_file = f'sample{sample_idx}.npy'
sample_path = os.path.join(base_dir, sample_file)
sample_data = np.load(sample_path, allow_pickle=True).item()

if attribute == 'x':
    selected_sample = sample_data['samples'][:, 0]
elif attribute == 'y':
    selected_sample = sample_data['samples'][:, 1]
elif attribute == 'z':
    selected_sample = sample_data['samples'][:, 2]
else:
    raise ValueError("Attribute must be 'x', 'y', or 'z'")

plt.figure(figsize=(10, 6))
plt.plot(selected_sample, label=f'Sample {sample_idx}', alpha=0.8)

plt.title(f'{system_type} V={V} parm{parm_idx} ({attribute})\n'
          f'β={sample_data["beta"]}, ε={sample_data["epsilon"]}')
plt.xlabel('Time Steps')
plt.ylabel(f'{attribute} Value')
plt.grid(True)
plt.legend()
plt.show()
```
