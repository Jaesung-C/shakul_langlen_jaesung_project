```python
import numpy as np
import matplotlib.pyplot as plt

system_type = 'SSS' 
V = np.inf
parm_idx = 9
attribute = 'x'
sample_idx = 2

base_dir = f'./data_set/{system_type}/euler/V{V}/parm{parm_idx}/'

sample_file = f'sample{sample_idx}.npy'
sample_path = os.path.join(base_dir, sample_file)

sample_data = np.load(sample_path, allow_pickle=True).item()

if attribute == 'x':
    selected_sample = sample_data['samples'][:, 0]
elif attribute == 'y':
    selected_sample = sample_data['samples'][:, 1]
elif attribute == 'z':
    selected_sample = sample_data['samples'][:, 2]
else:
    raise ValueError("Attribute must be 'x', 'y', or 'z'")

plt.figure(figsize=(10, 6))
plt.plot(selected_sample, label=f'Sample {sample_idx}', alpha=0.8)

plt.title(f'{system_type} V={V} parm{parm_idx} ({attribute})\n'
          f'β={sample_data["beta"]}, ε={sample_data["epsilon"]}')
plt.xlabel('Time Steps')
plt.ylabel(f'{attribute} Value')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

system_type = 'Bosc' 
V = np.inf
parm_idx_range = range(1, 12)  # You can modify this range as needed (e.g., 1 to 30)
attribute = 'x'
sample_idx = 1

base_dir_template = './data_set/{}/euler/V{}/parm{}/'

plt.figure(figsize=(10, 6))  # Create a single figure

for parm_idx in parm_idx_range:
    base_dir = base_dir_template.format(system_type, V, parm_idx)
    sample_file = f'sample{sample_idx}.npy'
    sample_path = os.path.join(base_dir, sample_file)

    sample_data = np.load(sample_path, allow_pickle=True).item()

    if attribute == 'x':
        selected_sample = sample_data['samples'][:, 0]
    elif attribute == 'y':
        selected_sample = sample_data['samples'][:, 1]
    elif attribute == 'z':
        selected_sample = sample_data['samples'][:, 2]
    else:
        raise ValueError("Attribute must be 'x', 'y', or 'z'")

    # Plot for each parm_idx on the same plot
    plt.plot(selected_sample, label=f'parm{parm_idx}', alpha=0.8)

# Finalize the plot
plt.title(f'{system_type} V={V} (attribute: {attribute}) for different parm_idx values\n')
plt.xlabel('Time Steps')
plt.ylabel(f'{attribute} Value')
plt.grid(True)
plt.legend()
plt.show()
```
